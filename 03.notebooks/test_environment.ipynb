```python
import os
import pandas as pd
from sqlalchemy import create_engine

print("--- PROBANDO BASE DE DATOS ---")
try:
    db_url = os.getenv("DATABASE_URL", "postgresql://ds_user:ds_password@db:5432/customer_churn")
    engine = create_engine(db_url)
    
    df_test = pd.DataFrame({'status': ['Conectado', 'Portátil'], 'codigo': [1, 2]})
    df_test.to_sql('test_conexion', engine, if_exists='replace', index=False)
    
    df_read = pd.read_sql('SELECT * FROM test_conexion', engine)
    print("✅ ¡ÉXITO! Conexión a PostgreSQL establecida.")
    print(df_read)
except Exception as e:
    print("❌ ERROR en Base de Datos:", e)
```
